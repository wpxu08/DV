```python
import numpy as np 
import matplotlib.pyplot as plt 
N = 50 
x = np.random.rand(N) 
y = np.random.rand(N) 
r = np.random.rand(N)
colors = np.random.rand(N) 
area = np.pi*(15*r)**2 # 0 to 15 point radiuses 
#plt.scatter(x, y) 
plt.scatter(x, y, s=area, c=colors, alpha = 0.5) 
plt.show() 
```

```python
import pandas as pd
df = pd.DataFrame(np.random.randn(100,4),columns = ['a','b','c','d'])
pd.plotting.scatter_matrix(df,figsize=(10,6),
                 marker = 'o',
                 diagonal='kde',
                 alpha = 0.5,
                 range_padding=0.1)
plt.show() 
```
